```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_g=pd.read_excel('zp_galuzi.xlsx')
display(df_g)
```

```python
#Ð´Ð¸Ð½Ð°Ð¼Ñ–ÐºÐ° Ð·Ð¼Ñ–Ð½ ÑÐµÑ€ÐµÐ´Ð½Ñ–Ñ… Ð·Ð°Ñ€Ð¾Ð±Ñ–Ñ‚Ð½Ñ–Ñ… Ð¿Ð»Ð°Ñ‚ Ð¿Ð¾ Ð³Ð°Ð»ÑƒÐ·ÑÐ¼
sns.set_style("whitegrid")
activity_types = df_g.columns[1:]
for activity_type in activity_types:
    plt.plot(df_g['Month'], df_g[activity_type], label=activity_type)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Revenue, UAH')
plt.show()
```
