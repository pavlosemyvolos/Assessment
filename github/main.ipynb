```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_g=pd.read_excel('zp_galuzi.xlsx')
display(df_g)
```

```python
#динаміка змін середніх заробітніх плат по галузям
sns.set_style("whitegrid")
activity_types = df_g.columns[1:]
for activity_type in activity_types:
    plt.plot(df_g['Month'], df_g[activity_type], label=activity_type)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Revenue, UAH')
plt.show()
```

```python
df_r=pd.read_excel('zp_region.xlsx')
display(df_r)
```

```python
#динаміка змін середніх заробітніх плат по регіонам виключаючи місто Київ
sns.set_style("whitegrid")
plt.figure(figsize=(15,7))
sns.lineplot(data=df_r)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
df_s=pd.read_excel('zp_stat.xlsx')
display(df_s)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
sns.lineplot(data=df_s, x='Month', y='Average by type of activity(men)', color='blue')
sns.lineplot(data=df_s, x='Month', y='Average by type of activity(women)', color='red')
plt.xticks(rotation=45)
plt.legend(labels=['Men', 'Women'])
plt.show()
```
